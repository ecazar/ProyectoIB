```python
import pandas as pd
import re
```

```python
path= './data/'
columnas = ['Score', 'Title', 'Review']
doc=pd.read_csv(path+'AmazonReview.zip', compression='zip', header=None, names=columnas)
doc
```

```python
def tokenizar(texto):
    # Convertir el texto a minúsculas
    texto = str(texto).lower()
    # expresión regular para encontrar todas las palabras (letras y números)
    tokens = re.findall(r'\b\w+\b', texto)
    return tokens
doc['combined'] = doc['Score'].astype(str) + " " + doc['Title'] + " " + doc['Review']
doc['tokens'] = doc['combined'].apply(tokenizar)
print(doc[['tokens']].head())
```
