```python
def clean_doc(doc):
    doc = re.sub(pattern=r'<.*?>', repl='', string=doc)
    doc = re.sub(pattern=r'[^A-Za-z0-9\-\']+', repl=' ', string=doc)
    doc = doc.lower()
    return re.sub(r'\s+', ' ', doc).strip()
def tokenizar(review):
    return review.split()
```

```python
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner", "tagger"])
porter_stemmer = PorterStemmer()
```

```python
def lemmatize_text(text):
    cleaned_text = clean_doc(text)
    doc = nlp(cleaned_text)
    lemmas = [token.lemma_ for token in doc if token.text.strip()]
    return " ".join(lemmas)
def lemmatize_text_opt(dataframe, column_name):
    cleaned_texts = dataframe[column_name].apply(clean_doc).tolist()
    lemmatized_results = []
    for doc in nlp.pipe(cleaned_texts, batch_size=5000):
        lemmas = [token.lemma_ for token in doc if token.text.strip()]
        lemmatized_results.append(" ".join(lemmas))
    return pd.Series(lemmatized_results, name=f'{column_name}')
def stem_text(text):
    cleaned_text = clean_doc(text)
    tokens = nltk.word_tokenize(cleaned_text)
    stems = [porter_stemmer.stem(token) for token in tokens]
    return " ".join(stems)
```

```python
df_lemma = lemmatize_text_opt(df_10, 'Review')
df_lemma
```
