```python
def clean_doc(doc):
    doc = re.sub(pattern=r'<.*?>', repl='', string=doc)
    doc = re.sub(pattern=r'[^A-Za-z0-9\-\']+', repl=' ', string=doc)
    doc = doc.lower()
    return re.sub(r'\s+', ' ', doc).strip()
```

```python
def lemmatize_text(text):
    cleaned_text = clean_doc(text)
    doc = nlp(cleaned_text)
    lemmas = [token.lemma_ for token in doc if token.text.strip()]
    return " ".join(lemmas)
```

```python
def stem_text(text):
    cleaned_text = clean_doc(text)
    tokens = nltk.word_tokenize(cleaned_text)
    stems = [porter_stemmer.stem(token) for token in tokens]
    return " ".join(stems)
```
