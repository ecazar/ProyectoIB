```python
import pandas as pd
import re
```

```python
path= './data/'
columnas = ['Score', 'Title', 'Review']
df=pd.read_csv(path+'AmazonReview.zip', compression='zip', header=None, names=columnas)
df
```

```python
def clean_doc(doc):
    doc = re.sub(pattern=r'<.*?>', repl='', string=doc)
    doc = re.sub(pattern=r'[^A-Za-z0-9\-\']+', repl=' ', string=doc)
    doc = doc.lower()
    return re.sub(r'\s+', ' ', doc).strip()
```

```python
def tokenizar(review):
    return review.split()
```

```python
import spacy
import nltk
from nltk.stem import PorterStemmer
import re
```

```python
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner", "tagger"])
porter_stemmer = PorterStemmer()
```

```python
def lemmatize_text(text):
    cleaned_text = clean_doc(text)
    doc = nlp(cleaned_text)
    lemmas = [token.lemma_ for token in doc if token.text.strip()]
    return " ".join(lemmas)
```

```python
def stem_text(text):
    cleaned_text = clean_doc(text)
    tokens = nltk.word_tokenize(cleaned_text)
    stems = [porter_stemmer.stem(token) for token in tokens]
    return " ".join(stems)
```

```python
df_lemma = df['Review'].apply(lemmatize_text)
df_lemma
```

```python
df_stem = df['Review'].apply(stem_text)
df_stem
```
